# 03. Correlation matrix heatmap

```python
# Create a correlation matrix heatmap based on the subset 
plt.matshow(df_subset.corr())
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_subset.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_subset.shape[1]), df_subset.columns, fontsize=14, rotation=90) # x axis labels
plt.yticks(range(df_subset.shape[1]), df_subset.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

### Because it is a little difficult to interpret, a correlation heatmap will be created with seaborn to compare.

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(12,12))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_subset.corr(), annot = True, ax = ax)
```

### Looking at the darkest coloured tiles, there are strong negative relationships between certain variables. In terms of 'Happiness Rank', the lowest number is 1, meaning that countries with lower numbers are the happier. The entire row of having variables cross 'Happiness Rank' produce moderate to strong correlations. Similarly, the 'Happiness Score' is just the inverse; instead of a strong negative relationship, it's a strong positive relationship as the 'Happiness Score' is what ranks the countries.

### When comparing the factors to the 'Happiness Score', there are many strong positive relationships, and the highest are wthin 'Economy (GDP per Capita) and 'Health (Life Expectancy)'. This indicates that residents are happieir because of their country's economy and living conditions.

# 06. Categorical plot

```python
# Create a histogram of the 'Happiness Score' variable 
sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
# Create a categorical variable that splits the 'Happiness Score' column into categories
df.loc[df['Happiness Score'] <= 4.5, 'Happiness Status'] = 'Unhappy country'
df.loc[(df['Happiness Score'] > 4.5) & (df['Happiness Score'] <= 6), 'Happiness Status'] = 'Medium happy country'
df.loc[df['Happiness Score'] > 6, 'Happiness Status'] = 'Happy country'
```

```python
# Check the value counts of the new column 'Happiness Status'
df['Happiness Status'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the status categories created above

sns.set(style = 'ticks')
cat_plot = sns.catplot(x = 'Continent', y = 'Happiness Score', hue = 'Happiness Status', data=df,  height = 5, aspect = 1.5,)
```

#### The categorical plot shows that many of the countries surveyed are within Africa and are considered unhappy. However, upon further research, Africa does have the most countries compared to the other continents so it makes sense. On the flip side, both North America and Australia have the least amount, but are all considered happy.
